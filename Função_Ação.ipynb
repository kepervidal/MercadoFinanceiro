```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
#Funçao que recebe um papel determinado, e cria um plot de cotação e volume, com dimensão e data, especificada pelo usuário
```

```python
 acao = input(str('Digite a ação:'))
 inicio = input(str('Digite a data de Inicio:'))
 fim = input(str('Digite a data de Fim:'))

def plot_acao(acao,inicio, fim):
    dados = yf.download(acao + '.SA', start= inicio, end= fim)
    fig = plt.figure(figsize= (12,6))  
    gs = fig.add_gridspec(nrows = 6, ncols = 1)
    ax1 = fig.add_subplot(gs[0:5, 0])
    ax2 = fig.add_subplot(gs[5,0])
   
    ax1.plot(dados.Close)
    ax2.bar(dados.index, dados.Volume)

plot_acao(acao, inicio, fim)
```
